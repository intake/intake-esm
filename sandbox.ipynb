```python
def _get_xarray_open_kwargs(data_format, xarray_open_kwargs=None, storage_options=None):
    xarray_open_kwargs = (xarray_open_kwargs or {}).copy()
    _default_open_kwargs = {
        'engine': 'zarr' if data_format == 'zarr' else 'netcdf4',
        'chunks': {},
        'backend_kwargs': {},
    }
    if not xarray_open_kwargs:
        xarray_open_kwargs = _default_open_kwargs
    else:
        xarray_open_kwargs = {**_default_open_kwargs, **xarray_open_kwargs}
    if (
        xarray_open_kwargs['engine'] == 'zarr'
        and 'storage_options' not in xarray_open_kwargs['backend_kwargs']
    ):
        xarray_open_kwargs['backend_kwargs']['storage_options'] = {} or storage_options
    return xarray_open_kwargs
```

```python
import pathlib

import intake
from ecgtools import Builder
from ecgtools.parsers.cesm import parse_cesm_timeseries
```

```python
root_path = pathlib.Path('tests/sample_data/tiny/').absolute()
```

```python
root_path = pathlib.Path('tests/sample_data/tiny/').absolute()
# where `tiny` is equivalent to `tiny = xr.DataArray(range(5), name='tiny').to_dataset()`

cat_builder = Builder(
    # Directory with the output
    root_path=root_path,
    # Depth of 1 since we are sending it to the case output directory
    depth=1,
    # Exclude the timeseries and restart directories
    exclude_patterns=["*/tseries/*", "*/rest/*"],
    # Number of jobs to execute - should be equal to # threads you are using
    njobs=5,
)

cat_builder
```

```python
cat_builder = cat_builder.build(parse_cesm_timeseries)
```
